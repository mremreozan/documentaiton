## The Mission

This time you'll have to work for Facebook! Facebook wants to develop a deep learning model that allows you to generate a text description from a photo. This is a big technical challenge because you have to combine your computer vision and NLP skills. 

### Must-have features

Your challenge will be to create descriptions that come closest to what a human could write. 

That is to say, your description should be a sentence and not just key words like "Dog, running, garden" but rather "A dog running in a garden".

### Steps
1. #### Image and Caption Dataset
The dataset has a pre-defined training dataset (6,000 images), development dataset (1,000 images), and test dataset (1,000 images)
    - [Download Flickr8k_dataset.zip (1 gb)](https://github.com/jbrownlee/Datasets/releases/download/Flickr8k/Flickr8k_Dataset.zip)
    - [Download Flickr8k_text.zip (2,2 mb)](https://github.com/jbrownlee/Datasets/releases/download/Flickr8k/Flickr8k_text.zip)
2. #### Prepare image data
As usual, it is necessary to prepare your images. (Put all the images at the right size, etc...)
3. #### Prepare text data
The dataset contains multiple descriptions for each photograph and the text of the descriptions requires some minimal cleaning.
4. #### Develop the deep learning model
Use a pre-formatted model to interpret image content and generate text. 
Take a look at Keras, Bert, OpenCV... You can choose your favorite tools.
5. #### Evaluate Model
Use the test and validation dataset to evaluate your model.  
[BLEU scores are used in text translation for evaluating translated text against one or more reference translations.](https://ariepratama.github.io/Introduction-to-BLEU-in-python/)!
6. #### Generate new captions
Use images taken at random on the net to generate new descriptions. 

## Evaluation criterias
1. Get a BLUE score of 0.50.
1. Use only pre-trained models.
